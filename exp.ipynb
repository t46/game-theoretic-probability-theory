```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("ticks")
sns.set_context("paper", 2, {"lines.linewidth": 2})
```

```python
def calc_capital(capital, bet, event, const=0.5):
    capital += bet * (event - const)
    return capital
```

```python
capital = 1
all_events = [0, 1]
total_trials = 1000
prob = 0
probs = []
for step in range(total_trials):
    bet = np.random.uniform(-1, 1)
    event = np.argmin(
        [
            calc_capital(capital, bet, 0),
            calc_capital(capital, bet, 1)
        ]
    )
    capital = calc_capital(capital, bet, event)
    prob += event / total_trials
    probs.append(prob)
print(prob)
```

```python
plt.plot(probs, color=(0.188, 0.333, 0.454))
plt.title('Probability of Coin-Toss Game')
plt.ylabel(r'$\frac{x_1 + x_2 + ... + x_n}{n}$')
plt.xlabel('Trial')
```
